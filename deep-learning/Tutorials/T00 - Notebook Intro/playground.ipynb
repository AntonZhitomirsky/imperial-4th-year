```python
import torch
import torch.nn.functional as F

max_pool_input = torch.tensor(
                    [[1,2,3,4,51],
                     [2,3,4,5,101],
                     [12,21,13,1,550],
                     [13,12,14,13,120],
                     [21,23,42,12,12]], dtype=torch.int64
                     )
max_pool_input
```

```python
# Forming input in the same dimensions as function requires (minibatch, in_channels, iH, iW)
# need to reshape it into a 4D tensor

max_pool_input = max_pool_input.reshape(-1,1,5,5)
max_pool_input
```

```python
F.max_pool2d(max_pool_input, kernel_size=2, stride=2, padding=1)
```

```python
# Import libraries
import matplotlib.pyplot as plt
import numpy as np
 
title='SIREN Activation'


# Creating vectors X and Y
x = np.linspace(-4, 4, 100)
y = F.siren(torch.Tensor(x))
 
fig = plt.figure(figsize = (5, 5))
# Create the plot
plt.plot(x, y,label='y=log_softmax(x)')
plt.grid()
plt.legend()
plt.title(title)
 
# Show the plot
plt.savefig(title)
```
